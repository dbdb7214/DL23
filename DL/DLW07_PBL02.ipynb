### 2. Create a function to quiz users on vocabulary words:

```python
#@markdown 🎧 Instruction #3
txt = "두 번째 단계에서는 단어 퀴즈를 만들기 위해 함수를 정희해 보겠습니다. 그 전에 단어가 임의순서로 나올 수 있도록 할 거예요. 그러면 지난 번에 배웠던 random이라는 모듈을 불러오면 되겠지요? 코드를 읽었다면 실행해 보세요."
tts(txt)
Audio("K-audio.mp3", autoplay = True)
```

```python
import random
```

```python
#@markdown 🎧 Instruction #
txt = "아래 코드는 함수를 만드는 거예요. 시작을 보면 d. e. f.라고 되어 있죠? 그리고 함수이름을 정하고 괄호 안에는 입력값이 무엇인지를 정해주었어요. 반드시 콜론으로 함수 시작은 끝나야 합니다. 그 아래 코드를 다 알 필요는 없어요. 다만, 무슨 일을 하는지는 봐야겠죠? 두 번째 줄에서는 단어와 의미 변수를 정하고, 거기에 렌덤하게 선택한 하나의 항목을 배당해 주었어요. 세 번째 줄에서는 사용자에게 선택된 단어의 뜻이 무엇이냐고 묻도록 화면에 print 해 주어요. 네 번째 줄은 사용자가 입력한 단어에서 혹시라도 있는 빈 공간을 없애고 모두 소문자로 바꾸어 줍니다. 다섯번째는 조건을 두어서, 답이 우리가 가진 의미를 소문자로 바꾼 것과 동일하면 정답이라고 화면에 출력해 주고, 그렇지 않으면 틀렸다고 답을 보내주는 거예요. 코드를 읽어본 후 아래 코드를 실행하세요."
tts(txt)
Audio("K-audio.mp3", autoplay = True)
```

```python
def quiz_vocabulary(vocab):
    word, meaning = random.choice(list(vocab.items()))
    print(f"What is the meaning of '{word}'?")
    answer = input().strip().lower()

    if answer == meaning.lower():
        print("Correct!")
    else:
        print(f"Wrong. The correct answer is '{meaning}'.")
```

3. Create a function to practice sentence construction:

```python
#@markdown 🎧 Instruction #
txt = "위에서 함수 사용하는 것을 보았는데, 아래 코드도 또 다른 함수를 정의하고 있습니다. 이번에는 이름을 문장연습이라 해 보았어요. 그리고 함수에 입력될 내용은 문장 중 하나겠죠. 이 함수는 앞에서 우리가 입력했던 3개의 문장 리스트가 있었잖아요? 그 중 하나를 랜덤하게 고르고, 문장을 스페이스 중심으로 나누어서 단어를 나열해요. 그리고 random.shuffle이라는 함수를 사용해서 단어 순서를 바꾸어 줍니다. 순서가 바뀐 단어를 나열해서 사용자에게 보여주면, 사용자가 영어에서 가능한 문장으로 바꾸어 입력하도록 요구합니다. 그리고 나서 원래 문장과 동일하면 정답이라고 말해주고, 다르면 틀렸다고 말한 후 정답을 보여주는 순서입니다. 코드를 읽어본 후 실행해 보세요."
tts(txt)
Audio("K-audio.mp3", autoplay = True)
```

```python
def practice_sentences(sentences):
    sentence = random.choice(sentences)
    words = sentence.split()
    random.shuffle(words)

    print("Reconstruct the sentence:")
    print(" ".join(words))
    answer = input().strip()

    if answer == sentence:
        print("Correct!")
    else:
        print(f"Wrong. The correct sentence is '{sentence}'.")
```

4. Create a main function to run the application:

```python
#@markdown 🎧 Instruction #
txt = "아래 코드는 main이라는 이름의 함수를 정해요. 이건 애플리케이션 화면이라 생각하시면 됩니다. 우리는 흐름만 이해하도록 합시다. 먼저 사용자가 3개의 옵션 중 하나를 선택하도록 화면에 보여줘요. 예를 들어, 1번은 단어퀴즈, 2번은 문장연습입니다. 3번은 끝내기예요. 선택한 번호를 입력으로 받아서 번호에 맞는 함수를 실행시키라는 코드입니다. 아래 코드를 대충 읽어보신 후 코드를 실행하세요."
tts(txt)
Audio("K-audio.mp3", autoplay = True)
```

```python
def main():
    while True:
        print("Choose an option:")
        print("1. Quiz vocabulary")
        print("2. Practice sentences")
        print("3. Quit")

        choice = input().strip()

        if choice == '1':
            quiz_vocabulary(vocab)
        elif choice == '2':
            practice_sentences(sentences)
        elif choice == '3':
            print("Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")
```

5. Call the main function to start the application:

```python
#@markdown 🎧 Instruction #
txt = "자, 이제 모든 함수를 정의했으니, 앱을 실행시켜 봅시다. 아래 코드는 실행시킴과 동시에 앱이 돌아가도록 해 주는 코드입니다. 이해하려고 하지 말고 그냥 실행시켜 보세요."
tts(txt)
Audio("K-audio.mp3", autoplay = True)
```

```python
if __name__ == "__main__":
    main()
```
